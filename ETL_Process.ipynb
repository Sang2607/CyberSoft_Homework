# **I. Import Data and Libraries**

```python
import pandas as pd
import xml.etree.ElementTree as ET
from datetime import datetime
endpoin = 'RawData/ETLFile'
log_file = 'log_file.txt'
target_file = 'transformed_date.csv'
```

# **II. Build Extract Function**

```python
def extract_from_CVS(fileToProcess):
    dataframe = pd.read_csv(fileToProcess)
    return dataframe

def extract_from_json(fileToProcess):
    dataframe = pd.read_json(fileToProcess)
    return dataframe

def extract_from_xml(fileToProcess):
    #B1: tạo khung
    dataframe = pd.DataFrame(columns=['name','height','weight'])
    #B2: gán dữ liệu vào
    tree = ET.parse(fileToProcess)
    root = tree.getroot()
    for person in root:
        name,height,weight = person.find('name').text,float( person.find('height').text),float(person.find('weight').text)
        Content = pd.DataFrame([{"name":name,'height':height,'weight':weight}])
        dataframe = pd.concat([dataframe,Content],ignore_index=True)
    return dataframe
    #cách 2
    #pd.read_xml(fileToProcess,parser='etree')
```
