## 1. Import Thư viện và data rỗng

```python
import json
import openpyxl 
warehouse = {}
file_name_json = 'warehouse.json'
file_name_excel = 'warehouse.xlsx'
```

## 2. Tạo hàm Warehouse_entry
*Ý nghĩa: Nhập mới hoặc cập nhật thêm số lượng sản phẩm đã có*

```python
def normalize_product_name(product_name): #normalize -> chuẩn hóa
    return product_name.strip().lower() #strip -> bỏ khoảng trắng đầu/đuôi => lower: viết thường lại

def warehouse_entry():
    print('\n--- INPUT GOODS DETAILS: ')
    product_name = normalize_product_name(input('Input product name: '))
    #Người dùng bỏ trống:
    if not product_name:
        print('The product name is not allowed to be empty!')
        return 
    try:
        inbound_quantity = int(input('Input inbound quantity: '))
        inbound_price = float(input('Input inbound unit price: '))
        if inbound_quantity < 0 or inbound_price < 0: 
            print('Quantity and price must be greater than 0')
            return #Nếu ko thỏa điều kiện IF đã định nghĩa -> ra khỏi HÀM
    except ValueError:
        print('Erros: Input values must be numeric!')
        return 
    #Xét giá trị trong warehouse:
    #TH1: Nếu warehouse đã có giá trị thì: price -> đổi theo price mới, quantity được cộng thêm!
    if product_name in warehouse:
        warehouse[product_name]['quantity'] += inbound_quantity
        warehouse[product_name]['price'] = inbound_price
        print(f'The quantity {inbound_quantity} units of {product_name} has been added!')
        print(f'The price of {product_name} has been updated to the new price is {inbound_price}') 
    #TH2 : Nếu warehouse chưa có, thì tạo:
    else:
        warehouse[product_name] = {"quantity" : inbound_quantity, "price": inbound_price}
        print(f"The new product {product_name} has been added successfully!")
```

## **3. Show warehouse**

```python
def show_warehouse():
    if not warehouse: #Nếu warehouse rỗng <> if warehouse: Nếu warehouse có:
        print("The warehouse is empty, please input new items.")
        return
    print(f"\n{'='*65}")
    print(f"{'Item list in Warehouse':^65}") #Canh giữa
    print(f"{'='*65}")
    print(f"{'Product Name':<25}{'Quantity':<12}{'Unit Price':<12}{'Total Amount':<15}")
    print(f"{'-'*65}")
    
    total_amount = 0 
    for product_name, details in warehouse.items():
        amount_per_item = details['quantity'] * details['price']
        total_amount += amount_per_item #total_amount = total_amount + amount_per_item
        print(f"{product_name:<25}{details['quantity']:<12}{details['price']:<12}{amount_per_item:<15}")
        
    print(f"Total Amount of Warehouse : {total_amount:,} USD")
    
```

## **4. Save lại thành file Json**

```python
def save_json_file(filename, data_json):
    with open(filename, 'w', encoding ='utf-8') as f:
        json.dump(data_json, f, ensure_ascii=False, indent =4)
```

#  **5. Load file from JSON file**

```python
# Biến cục bộ: Có giá trị trong 1 hàm
# Biến toàn cục: Có giá trị trong cả 1 session
def load_file_json(filename_json):
    global warehouse #biến toàn cục
    try:
        with open(filename_json, 'r', encoding='utf-8') as f:
            warehouse = json.load(f) #biến cục bộ
        print(f"The file {filename_json} has been loaded successfully!")
        print(warehouse)
    except FileNotFoundError:
        print("No json source file. Warehouse file is not found!")
```

# **6. Save file to Excel file:**


```python
def export_excel(file_name_excel):
    if not warehouse:
        print("Empty Warehouse, no data to export to Excel!")
        return
    #Tạo ra Workbook và Worksheet:
    wb = openpyxl.Workbook() #Tạo mới file Excel
    ws = wb.active #Nhấp đúp hoặc enter để file Open
    #Headers:
    headers = ['Product Name', 'Quantity', 'Unit Price', 'Line Total']
    ws.append(headers)
    
    #Đưa data vào các dòng sau Headers:
    for product_name, details in warehouse.items():
        line_total = details['price'] * details['quantity']
        row_data = [product_name, details['quantity'], details['price'], line_total]
        ws.append(row_data)
    wb.save(file_name_excel)
    print(f'Export successfully to {file_name_excel}')
    
```

```python
# Bước 1: Nhập data:
warehouse_entry()
```

```python
# Bước 2: Hiển thị các giá trị sản phẩm đã dc nhập vào
show_warehouse()
```

```python
# Bước 3: Lưu lại thành File JSON
save_json_file(file_name_json, warehouse)
```

```python
# Bước 4: Đọc lại giá trị của file json
load_file_json(file_name_json)
```

```python
# Bước 5: Xuất ra file Excel:
export_excel(file_name_excel)
```
