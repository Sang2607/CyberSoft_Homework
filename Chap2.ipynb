```python

XuatSac,Gioi,Kha,Tb  = 9,8, 6.5,5
score = int(input())
match score:
    case s if s >= XuatSac:
        print("Xuất sắc")
    case s if s >= Gioi:
        print("Giỏi")
    case s if s >= Kha:
        print("Khá")
    case s if s >= Tb:
        print("Trung bình")
    case _:
        print("Yếu")
```

```python
from Common.FunctionHandle import inputInt,inputStr,inputFloat
import sys
def Stop(): 
    print("kết thúc")
    sys.exit()
cardValid = True if inputStr("Nhập vào tình trạng thẻ: Y/N") == "Y" else Stop()
isBlock = True if inputStr("Thẻ có bị khóa không: Y/N") == "Y" else Stop()
fraud = True if inputStr("Xác nhận giao dịch này ko gian lận?: Y/N") == "Y" else Stop()
limit = inputFloat("Nhập hạn mức: ")
amount = inputFloat("Nhập số tiền thanh toán: ")

def Process():
    global limit,amount
    print("Hợp lệ")
    soTienConLai = limit - amount
    if(soTienConLai):
        print(f"Số tiền còn lại: {limit - amount}")
    else:
        print("Không hợp lệ")
Process() if cardValid and isBlock and fraud else print("Từ chối")

```

```python
from Common.FunctionHandle import inputInt

def valueInput():
    DoanhTHu = inputInt("Nhập doanh thu: ")
    ChiPhi = inputInt("Nhập chi phí: ")
    TinhLoiLo(DoanhTHu,ChiPhi)
def TinhLoiLo(DT,CP):
    loiNhuan = DT - CP;
    if loiNhuan == 0:
        print("Hòa vốn")
    else:
        print(f"Lãi: số tiền lãi: {loiNhuan}") if loiNhuan > 0 else print(f"Lỗ: số tiền lỗ: {loiNhuan}")

if __name__== "__main__":
    valueInput()
```

```python
from Common.FunctionHandle import inputInt

def InputValue():
    ThuNhap = inputInt("Nhập thu nhập: ")
    TinhThue(ThuNhap)
def TinhThue(TN = 0):
    MucThue = {
        10_000_000: 0.05,
        20_000_000: 0.1,
    }
    key = next((i for i in MucThue if TN <= i), False) #sử lý vòng lập xong rồi tiếp tục sử lý tuần tự sau dấu ,
    if(key):
        Thue_phai_nop = MucThue.get(key) * TN
        print(f"Thuế phải nộp:{Thue_phai_nop}")
    else:
        Thue_phai_nop = 0.15 * TN
        print(f"Thuế phải nộp:{Thue_phai_nop}")

InputValue()
```
