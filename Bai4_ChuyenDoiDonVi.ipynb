```python
def InputValue():
    soKnuts = inputInt()
    return Process(soKnuts)
def Process(K):
    galleon = K // (17 * 29)
    soDu = K % (17 * 29)
    sickle  = soDu // 29
    soDu = soDu % 29
    knut = soDu
    print(f"galleon: {galleon}"
          ,f"sickle: {sickle}"
          ,f"knut: {knut}"
          ,sep = "\n")
    
#output
if __name__ == "__main__":
    InputValue()
```
