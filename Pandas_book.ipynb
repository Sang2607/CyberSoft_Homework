# **V. Type String in Series**

```python
names = pd.Series(['Pomeray, CODY ',' Wagner; Jarry','smith, Ray'])
names = names.astype("string")
names
```

```python
names= names.str.replace(";",",")
names
```

```python
#Lấy độ dài
names.str.len()
```

```python
#Bỏ khoản trắng đầu đuôi
names = names.str.strip()
names
```

```python
#Viết Hoa all
names.str.upper()
names.str.lower()
```

```python
# pomeray, cody => [pomeray, cody]
#Xác định dấu hiện phân cách (delimiter) ', '
names = names.str.split(', ')
```

```python
#Đảo chuỗi
names = pd.Series([i[::-1] for i in names])
names = [' '.join(i) for i  in names]
names
```

# **VI. Date Type in Pandas**

```python
dateR = pd.period_range('01/01/2026',freq='30d',periods=6)
dateR
```

```python
#Lấy ngày đầu tiên
dateR.min()
```

```python
#Lấy ngày gần nhất
dateR.max()
```

```python
#Biến thành Dataqframe:
date_df = pd.DataFrame(data=dateR,columns=['sample_date'])
date_df
```

```python
#Lấy ngày trước và sau của ô hiện tại
date_df['prev_date'] = date_df['sample_date'].shift(periods=1)
date_df
```

```python
date_df['next_date'] = date_df['sample_date'].shift(periods=-1)
date_df
```

```python
#Cách 1
date_df['diff_date'] = date_df['next_date'] - date_df['sample_date']
#Cách 2
date_df["date_diff"] = date_df["sample_date"].diff(periods=1)

date_df

```

```python
#Tìm ngày đầu tiên trong tháng
date_df['fist_day_month'] = date_df['sample_date'].values.astype("datetime64[M]")
date_df
```

```python
#Tìm thứ trong tuần:
date_df = pd.DataFrame(data=dateR,columns=['sample_date'])
date_df['sample_date'] = date_df['sample_date'].dt.to_timestamp()
date_df['sample_date'].dt.day_name()
```
