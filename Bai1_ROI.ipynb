```python
from Common.FunctionHandle import inputInt
from Common.ExcelHandle import SaveFile
```

```python
def InputValue():
    investmentCosts = inputInt()
    expectedIncome = inputInt()
    print(
        f"Chi phí đầu tư: {investmentCosts:,}"
        ,f"Thu nhập ròng: {expectedIncome:,}"
        , sep = "\n"
        )
    return ProcessRoi(investmentCosts,expectedIncome)
def ProcessRoi(chiPhi,thuNhap):
    result = (chiPhi/thuNhap)*100
    print(f"Chỉ Số Hiệu Quả (ROI): {result}")
    
    #Lưu data vào excel
    FileName = "Files/Bai1_ROI.xlsx"
    Title = "Calculating_ROI"
    NestList = []
    header = ["Chi phí đầu tư","Thu nhập ròng","Chỉ Số Hiệu Quả (ROI)"]
    NestList.append(header)
    NestList.append([chiPhi,thuNhap,result])
    SaveFile(FileName,Title,NestList)
    NestList.clear()

def OutputROI():
    InputValue()

if __name__ == "__main__":
    OutputROI()
```
