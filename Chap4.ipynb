```python
num_list = [int(num.strip()) for num in input("Nhập vào dãy số nguyên: ").split(',')]
for num in num_list:
    sum = 0
    for i in range(2, num):
        if num % i == 0:
            sum +=1
    if sum ==0:
        status = "Số nguyên tố"
    else:
        status = "Không phải số nguyên tố"
    print(f"{num} : {status}")
```
