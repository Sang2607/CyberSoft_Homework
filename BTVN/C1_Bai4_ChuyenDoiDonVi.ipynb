```python
import bootstrap
from Common.FunctionHandle import inputInt
from Common.ExcelHandle import SaveFile
from Common.Decorator import handle_exception
```

```python
FileName = "Files/Bai4_ChuyenDoi.xlsx"
Title = "KeTien"
ListData = []
soKnuts = 0
```

```python
@handle_exception
def InputValue():
    global soKnuts
    soKnuts = inputInt("Nhập tổng số Knuts: ")
    return Process(soKnuts)
def Process(K):
    global ListData
    G,S = 29, 17
    galleon = K // (S * G)
    soDu = K % (S * G)
    sickle  = soDu // G
    soDu = soDu % G
    knut = soDu
    print(f"galleon: {galleon}"
          ,f"sickle: {sickle}"
          ,f"knut: {knut}"
          ,sep = "\n")
    ListData = AppendList(galleon,sickle,knut)
def AppendList(galleon,sickle,knut):
    NestList = []
    NestList.append(["Tổng Knut", soKnuts])
    NestList.append(["Loại đơn vị", "Số lượng"])
    NestList.append(["Số Galleon", galleon])
    NestList.append(["Số Sickle", sickle])
    NestList.append(["Số dư Knut", knut])
    return NestList
#output
if __name__ == "__main__":
    InputValue()
    SaveFile(FileName,Title,ListData,"Bạn có muốn xuất excel không?",removeHeder=False)

```
