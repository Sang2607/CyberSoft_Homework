```python
import bootstrap
from Common.FunctionHandle import inputInt,inputFloat
from Common.ExcelHandle import SaveFile
```

```python
def InputValue():
    soTienVay = inputInt("Nhập số tiền vay: ")
    soThang = inputInt("Nhập thời hạn vay: ")
    laiSuatNam = inputFloat("Nhập lãi suất vay: ")
    print(
        f"Sồ tiền vay: {soTienVay:,}"
        ,f"Thời hạn: {soThang:,}"
        ,f"Lãi suất(năm): {laiSuatNam:,}"
        , sep = "\n"
        )
    return Process(soTienVay,soThang,laiSuatNam)
def Process(P,n,rate):
    NestList = []
    r = rate/12
    for i in range(1,n + 1):
        result = (P*r*(1+r)**n) / ((1+r)**n - 1)
        match i:
            case 1:
                NestList.append(["Sồ tiền vay","Thời hạn","Lãi suất/năm","",f"Tháng {i}",result])
            case 2:
                NestList.append([P,n,rate,"",f"Tháng {i}",result])
            case _:
                NestList.append(["","","","",f"Tháng {i}",result])
        print(f"Tháng {i}: {result:,}")
    
    #Lưu data vào excel
    FileName = "Files/Bai2_DongTienVay.xlsx"
    Title = "DongTien"
    SaveFile(FileName,Title,NestList,"Bạn có muốn xuất excel không?",removeHeder = False)
    NestList.clear()

#output
if __name__ == "__main__":
    InputValue()
```
