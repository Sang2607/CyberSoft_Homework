```python
nums = int(input())
Total = 1
for i in range(1,nums +1):
    Total = Total * i
    print(Total)

def deQuy(nums):
    if nums == 0:
        return 1
    return nums * deQuy(nums - 1)
print(deQuy(nums))
    

```

```python
nums_list = input().split(" ")

numMax = 0
for i in nums_list:
    num = int(i.strip()) #strip: loại bỏ khoản trống như trim
    if numMax < num:
        numMax = num
print(numMax)
```
