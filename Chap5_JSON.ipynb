```python
cart_data = """{ "items": [ {"name": "Áo thun", "price": 200000, "quantity": 2}, {"name": "Quần Jean", "price": 450000, "quantity": 1} ] }"""
cart = json.loads(cart_data)
def  calculate_total(cart_json,discount_percent):
    total = 0
    for i in cart_json["items"]:
        total += (i["price"] * i["quantity"])

    print(f"{total*(1 - discount_percent):,}")

discount = float(input())
calculate_total(cart,discount)
```

Bài 2: Viết hàm split

```python
order_data = """ { "products": 
[ 
{"shop_id": "SHOP_A", "name": "Chuột máy tính", "price": 150000}, 
{"shop_id": "SHOP_B", "name": "Sách kỹ năng", "price": 85000}, 
{"shop_id": "SHOP_A", "name": "Bàn phím", "price": 300000} ] } """
data  = json.loads(order_data)["products"]

new_array = []
group_shop = {}
for item in data:
    shop_id = item["shop_id"]

    if shop_id not in group_shop:
        group_shop[shop_id] = []
    group_shop[shop_id].append({
                "name": item["name"],
                "price": item["price"]
            })

 

print(group_shop)
```
